# Análise Exploratória de Dados - Vehicles
Este notebook apresenta uma análise inicial do arquivo `vehicles_us.csv`.


```python
import pandas as pd
import plotly.express as px
```

```python
car_data = pd.read_csv('../vehicles_us.csv')
car_data.head()
```

```python
car_data.info()
```

```python
car_data.describe()
```

```python
fig = px.histogram(car_data, x='price', nbins=30, title='Distribuição dos preços')
fig.show()
```

```python
fig = px.scatter(car_data, x='odometer', y='price', color='condition', title='Preço x Odômetro')
fig.show()
```
